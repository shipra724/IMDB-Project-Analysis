```python
# open webdriver chrome
driver=webdriver.Chrome("chromedriver.exe")
# maximize the automated chrome window
driver.maximize_window()
# open amazon on the the automated chrome window
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
```

```python
# scraping S.no 
s_number=[]
start_page=0
end_page=34

for page in range(start_page,end_page):
    num_tags=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
    for i in num_tags:
        s_number.append(i.text)
    
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(20)
    
print(s_number)
```

```python
# scraping votes
votes=[]
start_page=0
end_page=34

for page in range(start_page,end_page):
    votes_tags=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in votes_tags:
          votes.append(i.text[6:].split('|')[0].replace('Votes :',''))
    
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(5)
    
print(votes)
```

```python
# scraping movie genre
gross_collection=[]
start_page=0
end_page=34
for page in range(start_page,end_page):
    gross_coll_tags=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in gross_coll_tags:
        gross_text=i.text.split('|')
        if len(gross_text) >=2:
            gross_collection.append(i.text.split('|')[1].replace('Gross:','').strip())
        else :
            gross_collection.append('-')
    
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(3)
    
print(gross_collection)
```

```python
#returning back to home page
s_page=34
e_page=0
for p in range(s_page,e_page,-1):
    pre_bttn=driver.find_element(By.XPATH,'//a[@class="lister-page-prev prev-page"]')
    pre_bttn.click()
    time.sleep(2)
```

```python
# scraping certification
certification=[]
start_page=0
end_page=34

for page in range(start_page,end_page):
    for i in range(50) :
        certification_tags=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[2]/div[3]/div[1]/div/div[3]/div/div['+str(i+1)+']/div[3]/p[1]/span[1]')
        certification_text = certification_tags.text
        if ' min' in certification_text :
            certification.append('-')
        else:
            certification.append(certification_text) 
    
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(5)
    
print(certification)
```

```python
#returning back to home page
s_page=34
e_page=0
for p in range(s_page,e_page,-1):
    pre_bttn=driver.find_element(By.XPATH,'//a[@class="lister-page-prev prev-page"]')
    pre_bttn.click()
    time.sleep(2)
```

```python
#scraaping metascore
metascore = []
start_page=0
end_page=34

for page in range(start_page,end_page):
    metascore_tags=driver.find_elements(By.XPATH,'//div[@class="ratings-bar"]')
    for i in metascore_tags:
        tempMetaScore = i.text.split('\n')
        if (len(tempMetaScore) == 2) :
            metascore.append(tempMetaScore[1])
        else:
            metascore.append('NA')
    
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(5)
    
```

```python
#returning back to home page
s_page=34
e_page=0
for p in range(s_page,e_page,-1):
    pre_bttn=driver.find_element(By.XPATH,'//a[@class="lister-page-prev prev-page"]')
    pre_bttn.click()
    time.sleep(2)
```

```python
#bifurcating directors into director1,director2
director1=[]
director2=[]

#bifurcating stars into star1,sar2,star3,star4
star1=[]
star2=[]
star3=[]
star4=[]

#bifurcating genre into star1,sar2,star3,star4
genre1=[]
genre2=[]
genre3=[]

start_page=0
end_page=34
for page in range(start_page,end_page):
    for i in range(50) :
        stars_directors_text=stars_tags=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[2]/div[3]/div[1]/div/div[3]/div/div['+str(i+1)+']/div[3]/p[3]').text.split(' | ')
        director_text = stars_directors_text[0]
        stars_text = stars_directors_text[1]
        if('Directors' in director_text) :
            directorAll = director_text.replace('Directors: ', '').split(',')
            director1.append(directorAll[0])
            director2.append(directorAll[1])
        else: 
            director1.append(director_text.replace('Director: ', ''))
            director2.append('-')
            
        if('Star:' in stars_text) :
            starsAll = stars_text.replace('Star: ','').split(',')
            star1.append(starsAll[0])
            star2.append('-')
            star3.append('-')
            star4.append('-')
        else :
            starsAll = stars_text.replace('Stars: ','').split(',')
            star1.append(starsAll[0])
            star2.append(starsAll[1].strip())
            if len(starsAll)>2 :
                star3.append(starsAll[2].strip())
            else :
                star3.append('-')
            if len(starsAll)>3 :
                star4.append(starsAll[3].strip())
            else :
                star4.append('-')
            
    movieGenreTags = driver.find_elements(By.XPATH,'//span[@class="genre"]')
    for i in movieGenreTags :
        genre_text = i.text
        if(',' in genre_text) :
            genreAll = genre_text.split(',')
            genre1.append(genreAll[0])
            genre2.append(genreAll[1].strip())
            if len(genreAll)>2 :
                genre3.append(genreAll[2].strip())
            else :
                genre3.append('-')
        else :
            genre1.append(genre_text)
            genre2.append('-')
            genre3.append('-')
        
    next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    next_button.click()
    time.sleep(5)
        
```
