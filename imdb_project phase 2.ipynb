```python
import sqlite3
import csv
import pandas as pd
```

```python
db=sqlite3.connect('IMDB_database.db')
```

```python
cursor=db.cursor()
```

```python
cursor.execute("CREATE TABLE IMDB_table (col1 TEXT, col2 TEXT, col3 TEXT, col4 TEXT, col5 TEXT, col6 TEXT, col7 TEXT, col8 TEXT);")
```

```python

with open('file1.csv', 'r', encoding='utf-8') as file:
    total_records = 0
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        # Retrieve the values from the row
        col1_value = row.get('S.No')
        col2_value = row.get('Movie Name')
        col3_value = row.get('Director Name_1')
        col4_value = row.get('Director Name_2')
        col5_value = row.get('Duration')
        col6_value = row.get('Year')
        col7_value = row.get('Ratings')
        col8_value = row.get('Metascore')

        # Execute the query with the values
        cursor.execute("INSERT INTO IMDB_table (col1, col2, col3, col4, col5, col6, col7, col8) VALUES (?, ?, ?, ?, ?, ?, ?, ?)",
                    (col1_value, col2_value, col3_value, col4_value, col5_value, col6_value, col7_value, col8_value))
        total_records += 1
db.commit()
print(total_records, 'records inserted')
```
